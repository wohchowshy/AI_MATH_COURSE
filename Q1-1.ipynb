```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

stock = pd.read_csv('stock.csv')
TCC = np.array(stock['end'][1000:1332])

yMax = max(TCC)
xMax = TCC.tolist().index(max(TCC))

plt.xlabel('Date')
plt.ylabel('End Price')
plt.suptitle('Historical stock price of TCC')
plt.annotate('Golbal max', xy=(xMax, yMax), xytext=(200, 44),
             arrowprops=dict(facecolor='black', shrink=0.05),)
plt.plot(TCC,'g')
```
