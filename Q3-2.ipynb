```python
%matplotlib inline

#比較'10個隨機點'和'10000個隨機點'的sine分布圖

import matplotlib.pyplot as plt
import numpy as np

#隨機產生10個[0,2pi]之間的數
x10 = np.random.rand(10)*np.pi*2
x10 = np.sort(x10)
y10 = np.sin(x10)
plt.xlim(0,2*np.pi)
plt.ylim(-1,1)
plt.plot(x10, y10, linewidth='2.0',label='10 random points')


#隨機產生100個[0,2pi]之間的數
x1000 = np.random.rand(1000)*np.pi*2
x1000 = np.sort(x1000)
y1000 = np.sin(x1000)
plt.plot(x1000, y1000, color='red', linewidth='1.0', linestyle='--', label='10000 random points')

plt.legend(loc='upper right')

```
