```python
# 簡單的matplot動畫
# 使用random函數
%matplotlib inline
import numpy as np 
import matplotlib.pyplot as plt
from matplotlib import animation

fig, ax = plt.subplots()

plt.xlim(0,10)
plt.ylim(0,10)
x = np.arange(0, 11, 1)
line, = ax.plot(x, np.random.random(len(x))*10, 'r-')

def animate(i):
    line.set_ydata(np.random.random(len(x))*10)
    return line,

def init():
    line.set_ydata(np.random.random(len(x))*10)
    return line,

ani = animation.FuncAnimation(fig=fig, func=animate, frames=10, 
                              init_func=init, interval=200, blit=False)

plt.show()


```
